## Visualizing Earnings Based on College Majors

Columns of the dataset:

* Rank - Rank by median earnings (the dataset is ordered by this column).
* Major_code - Major code.
* Major - Major description.
* Major_category - Category of major.
* Total - Total number of people with major.
* Sample_size - Sample size (unweighted) of full-time.
* Men - Male graduates.
* Women - Female graduates.
* ShareWomen - Women as share of total.
* Employed - Number employed.
* Median - Median salary of full-time, year-round workers.
* Low_wage_jobs - Number in low-wage service jobs.
* Full_time - Number employed 35 hours or more.
* Part_time - Number employed less than 35 hours.




```python
import pandas as pd
import matplotlib.pyplot as plt

#plots displayed inline
%matplotlib inline
```

```python
recent_grads = pd.read_csv('recent-grads.csv')
rg = recent_grads
raw_data_count = len(rg.index)



print(rg.iloc[0])
print(rg.head())
print(rg.tail())

rg = rg.dropna()
clean_data_count = len(rg.index)


print('Raw Data Count:',raw_data_count)
print('Clean Data Count:', clean_data_count)
```

```python
rg.plot(x='Sample_size', y='Median', kind='scatter', title='Sample_size vs. Median')
```

```python
rg.plot(x='Sample_size', y='Unemployment_rate', kind='scatter', title='Sample_size vs. Unemployment_rate')
```

```python
rg.plot(x='Full_time', y='Median', kind='scatter', title='Full_time vs. Median')
```

```python
rg.plot(x='ShareWomen',y='Unemployment_rate', kind='scatter')
```

```python
rg.plot(x='Men',y='Median',kind='scatter',title='Men vs. Median')
```

```python
rg.plot(x='Women', y='Median', kind='scatter', title="Women vs. Median" )
```

Do students in more popular majors make more money?
Do students that majored in subjects that were majority female make more money?
Is there any link between the number of full-time employees and median salary?

The answers require further exploration through other types of graphs such as histograms.

### Distribution exploration (Histograms)

```python
cols=['Sample_size', "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate"]

fig = plt.figure(figsize=(5,12))

for c in range(1,5):
    ax = fig.add_subplot(4,1,c)
    ax = rg[cols[c]].plot(kind='hist', rot=40)
```

```python
cols=['Sample_size', "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate","Men","Women"]

fig = plt.figure(figsize=(5,12))
for c in range(4,8):
    ax = fig.add_subplot(4,1,c-3)
    ax = rg[cols[c]].plot(kind='hist',rot=40)
```
